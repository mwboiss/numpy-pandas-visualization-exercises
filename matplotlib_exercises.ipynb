```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
import pandas as pd
```

```python
x = range(-5,5)
y = [n**2 - n + 2 for n in x]

plt.plot(x,y)

plt.text(0,0.75, '0,0', fontsize = 10, color = 'g')

plt.show()
```

```python
x = range(0,20)
y = [n**(1/2) for n in x]

plt.plot(x,y)

plt.text(0.25,0, '0,0', fontsize = 10, color = 'g')
plt.title('$\sqrt{x}$')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = range(-20,20)
y = [n**3 for n in x]

plt.plot(x,y)

plt.text(0,100, '0,0', fontsize = 10, color = 'g')
plt.title('$x^3$')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = range(-10,10)
y = [2**n for n in x]

plt.plot(x,y)

plt.text(0,5, '0,0', fontsize = 10, color = 'g')
plt.title('$2^X$')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = range(0,20)
y = [1 / (n + 1) for n in x]

plt.plot(x,y)

plt.title('$\frac{1}{x+1}$')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = range(0,5)
cube_x = [n**3 for n in x]
one_over_x_plus_one = [1 / (n + 1) for n in x]
sqrt_x = [n**(1/2) for n in x]
two_to_the_x = [2**n for n in x]

plt.figure(figsize = (15,10))

plt.subplot(221)
plt.plot(x,cube_x)
plt.title('$x^3$')

plt.subplot(222)
plt.plot(x,one_over_x_plus_one)
plt.title('$1/(x + 1)$')

plt.subplot(223)
plt.plot(x,sqrt_x)
plt.title('$sqrt(x)$')

plt.subplot(224)
plt.plot(x,two_to_the_x)
plt.title('$2^x$')


plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = range(0,5)
cube_x = [n**3 for n in x]
one_over_x_plus_one = [1 / (n + 1) for n in x]
sqrt_x = [n**(1/2) for n in x]
two_to_the_x = [2**n for n in x]

plt.figure(figsize = (15,10))
plt.plot(x,cube_x, color = 'g', label = 'X^3')
plt.plot(x,one_over_x_plus_one, color = 'b', label = '1/(X + 1)')
plt.plot(x,sqrt_x, color = 'r', label = 'Square Root of X')
plt.plot(x,two_to_the_x, color = 'c', label = 'Two to the X')

plt.title('Maths on Graphs')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
