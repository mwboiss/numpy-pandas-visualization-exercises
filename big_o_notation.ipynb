```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(1,4)
O_one = [0 * n + 1 for n in x]
O_log_n = [math.log(n) for n in x]
O_n = [n for n in x]
O_n_log_n = [n * math.log(n) for n in x]
O_n_squared = [n**2 for n in x]
O_two_to_the_n = [2**n for n in x]
O_n_factorial = [math.factorial(n) for n in x]
O_n_to_the_n = [n**n for n in x]

plt.figure(figsize = (15,10))
plt.plot(x,O_one, color = 'b', label = 'O(1)')
plt.plot(x,O_log_n, color = 'g', label = 'O(log n)')
plt.plot(x,O_n, color = 'r', label = 'O(n)')
plt.plot(x,O_n_log_n, color = 'c', label = 'O(n log n)')
plt.plot(x,O_n_squared, color = 'm', label = 'O(n^2)')
plt.plot(x,O_two_to_the_n, color = 'y', label = 'O(2^n')
plt.plot(x,O_n_factorial, color = 'k', label = 'O(n!)')
plt.plot(x,O_n_to_the_n, color = 'b', label = 'O(n^n)')


plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
plt.legend()
plt.show()
```
